#Method of images

"The method of images (or method of mirror images) is a mathematical tool for solving differential equations, in which the domain of the sought function is extended by the addition of its mirror image with respect to a symmetry hyperplane. As a result, certain boundary conditions are satisfied automatically by the presence of a mirror image, greatly facilitating the solution of the original problem". [Definition from Wikipedia](http://en.wikipedia.org/wiki/Method_of_images)

**Question**: can we get flow around a straight wall? Yes, we can!

In our case the method of images is simply this: put a singularity near a "wall" by adding an "image" singularity on the other side of the wall. Sometimes, this is also referrered to as *aerodynamic interferences*. 


##Source near a plane wall

If a source is located near a wall, the flow pattern will be distorted by the wall. Imagine a source located at $y=y_\text{source}$ next to a wall at $y=0$. The boundary condition at the wall requires the flow to be tangent there, for a horizontal wall, $v=0$. The effect of the wall on the source flow is identical to placing another source (an image) of equal strength, located at $y=-y_\text{source}$.

We'll use classes for this lesson, go to [lesson07](http://nbviewer.ipython.org/github/barbagroup/AeroPython/blob/master/lessons/07_Lesson07_methodOfImages.ipynb) for a brief introduction to classes. 

```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Mesh-grid
N = 50                                   # Number of points in each direction
x_start, x_end = -2.0, 2.0               # x-direction boundaries
y_start, y_end = -1., 1.                 # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
strength_source = 1.0                    # strength of the source         
x_source, y_source = 0.0, 0.5            # location of the source
```

```python
strength_vortex = 1.0                  # strength of the vortex
x_vortex, y_vortex = 0.0, 0.5          # location of the vortex
```

```python
class Source:
    """Contains information related to a source (or sink)."""
    def __init__(self, strength, x_s, y_s):
        """Initializes the singularity.
        
        Arguments
        ---------
        strength: strength of the singularity.
        x_s, y_s: coordinates of the singularity.
        """
        self.strength = strength
        self.x_s, self.y_s = x_s, y_s
    
    def velocity(self, X, Y):
        """Computes the velocity field generated by the singularity.
        
        Arguments
        ---------
        X, Y: mesh grid.
        """
        self.u = self.strength/(2*numpy.pi)*(X-self.x_s)/((X-self.x_s)**2+(Y-self.y_s)**2)
        self.v = self.strength/(2*numpy.pi)*(Y-self.y_s)/((X-self.x_s)**2+(Y-self.y_s)**2)
    
    def stream_function(self, X, Y):
        """Computes the stream-function generated by the singularity.
        
        Arguments
        ---------
        X, Y: mesh grid.
        """
        self.psi = self.strength/(2*numpy.pi)*numpy.arctan2((Y-self.y_s), (X-self.x_s))
```

```python
# creates a source (object of class Source) 
source = Source(strength_source, x_source, y_source)

# computes the velocity field and the stream-function on the mesh grid
source.velocity(X, Y)
source.stream_function(X, Y)
```

```python
# creates the image (note we use -y_source) of the source and computes velocity and stream-function
source_image = Source(strength_source, x_source, -y_source)

# computes the velocity field and the stream-function on the mesh grid
source_image.velocity(X, Y)
source_image.stream_function(X, Y)
```

Note that now we have the attributes u and v, and also psi in each case, so we can invoke superposition and get the velocity and the streamfunction for the whole system. 

```python
# superposition of the source and its image
u_source = source.u + source_image.u
v_source = source.v + source_image.v
psi_source = source.psi + source_image.psi
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#The source and the image
pyplot.scatter(source.x_s, source.y_s, color='r', s=80, marker='o')
pyplot.scatter(source_image.x_s, source_image.y_s, color='r', s=80, marker='D')

#streamfunction
pyplot.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B')

#dash line in the x-axis simulating the wall
pyplot.axhline(0., color='k', linestyle='--', linewidth=4);
```

##Vortex near a plane wall

Following the same idea that in the case of the source, we write another class called `Vortex`.

```python
class Vortex:
    """Contains information related to a vortex."""
    def __init__(self, strength, x_v, y_v):
        """Initializes the vortex.
        
        Arguments
        ---------
        strength: strength of the vortex.
        x_v, y_v: coordinates of the vortex.
        """
        self.strength = strength
        self.x_v, self.y_v = x_v, y_v
        
    def velocity(self, X, Y):
        """Computes the velocity field generated by a vortex.
        
        Arguments
        ---------
        X, Y: mesh grid.
        """
        self.u = +self.strength/(2*numpy.pi)*(Y-self.y_v)/((X-self.x_v)**2+(Y-self.y_v)**2)
        self.v = -self.strength/(2*numpy.pi)*(X-self.x_v)/((X-self.x_v)**2+(Y-self.y_v)**2)
        
    def stream_function(self, X, Y):
        """Computes the stream-function generated by a vortex.
        
        Arguments
        ---------
        X, Y: mesh grid.
        """
        self.psi = -self.strength/(4*numpy.pi)*numpy.log((X-self.x_v)**2+(Y-self.y_v)**2)
```

```python
# creates a vortex and computes velocity and stream-function
vortex = Vortex(strength_vortex, x_vortex, y_vortex)
vortex.velocity(X, Y)
vortex.stream_function(X, Y)

# creates the image of the vortex and computes velocity and stream-function
vortex_image = Vortex(-strength_vortex, x_vortex, -y_vortex)
vortex_image.velocity(X, Y)
vortex_image.stream_function(X, Y)
```

```python
# superposition of the vortex and its image
u_vortex = vortex.u + vortex_image.u
v_vortex = vortex.v + vortex_image.v
psi = vortex.psi + vortex_image.psi
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#The vortex and the image
pyplot.scatter(vortex.x_v, vortex.y_v, color='maroon', s=80, marker='o')
pyplot.scatter(vortex_image.x_v, vortex_image.y_v, color='maroon', s=80, marker='D')

#streamfunction
pyplot.streamplot(X, Y, u_vortex, v_vortex, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B')

#dash line in the x-axis simulating the wall
pyplot.axhline(0., color='k', linestyle='--', linewidth=4);
```
