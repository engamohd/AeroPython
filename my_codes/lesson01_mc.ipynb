This notebook is my own version of the lesson01 of Aeropython. Here we will visualize the stramlines corresponding to source and sink.

```python
#Importing what we need

import numpy
from matplotlib import pyplot
from math import *  

#The next line configures the plotting module to output plots inside the ipynb 
%matplotlib inline                 
```

Let's define theset of point where the components of the velocity will be computed. We will use an evenly spaced cartesian grid.

```python
N = 50                                     # Number of point in each direction

x_start, x_end = -2.0 , 2.0                # x-boundaries (length 4)
y_start, y_end = -1.0 , 1.0                # y-boundaries (length 2)
 
x = numpy.linspace(x_start, x_end, N)      # 1D array in x-direction equally spaced and with N points
y = numpy.linspace(y_start, y_end, N)      # 1D array in y-direction equally spaced and with N points
```

```python
#Let's generate the mesh grid using x and y

X, Y = numpy.meshgrid(x, y)  
```

```python
# And plotting the points

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X, Y, s=10, color='r', marker='o', linewidth=0);
```

Now let's calculate the velocity vector corresponding to a source flow and then we will plot the streamlines .

###Source

To explain what a is a source, we can imagine a "point" where the fluid is flowing out, uniformly. In other words, all the streamlines radiate from this point as straight lines and the radial velocity decreases with the distance from the source point. In 2D, we can write the velocity components as:

$$u_r(r,\theta) = \frac{\sigma}{2\pi r} \text{,} \qquad u_{\theta}(r,\theta)=0$$

where $\sigma$ is the strength of the source.

In Cartesian coordinates, the velocity field $(u,v)$ at position $(x,y)$ corresponding to a source of strength $\sigma$ located at $(x_\text{source},y_\text{source})$ is given by:

\begin{equation}
    	\begin{split}
        u = u_r\;cos(\theta) &= u_r \frac{x-x_\text{source}}{[\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2]^{1/2}}\\\\ &=  \frac{\sigma}{2\pi}\frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}
        \end{split}
\end{equation}

and

\begin{equation}
    	\begin{split}
        u = u_r\;sin(\theta) &= u_r \frac{y-y_\text{source}}{[\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2]^{1/2}}\\\\ &=  \frac{\sigma}{2\pi}\frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}
        \end{split}
\end{equation}



```python
# Let's compute the velocity field on the mesh grid

strength_source = 5.0
x_source, y_source = -1.0, 0.0         # source location

# u is the x-component and v is the y-component

u_source = strength_source/(2*pi) * (X-x_source) / ((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*pi) * (Y-y_source) / ((X-x_source)**2 + (Y-y_source)**2)
```

```python
#Now plotting the streamlines using streamplot() and scatter for the source

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_source, y_source, s=80, color='r', marker='o', linewidth=0)
pyplot.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B');
```

###Sink

We can think in a sink as a source with a negative strength. Now, instead of have radiating streamlines from a single point, we will have streamlines converging to a single point.

```python
# Let's compute the velocity field on the mesh grid

#strength_sink = - strength_source      #we can use other value, not necessary to have the same value
strength_sink = -25
x_sink, y_sink = 1.0, 0.0               # source location

# u is the x-component and v is the y-component

u_sink = strength_sink/(2*pi) * (X-x_sink) / ((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*pi) * (Y-y_sink) / ((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
#Now plotting the streamlines using streamplot() and scatter for the source

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_sink, y_sink, s=80, color='g', marker='o', linewidth=0)
pyplot.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B');
```

###Source-sink pair

Using superposition, (we can do it cause the linearity of the problem), we are going to build a new solution of the potential field, source-sink pair.

```python
#Summing the source and sink velocities we get the pair velocity

u_pair = u_source + u_sink
v_pair = v_source + v_sink
```

```python
#Now plotting the streamlines using streamplot() and scatter for the source

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_sink, y_sink, s=80, color='g', marker='o', linewidth=0)
pyplot.scatter(x_source, y_source, s=80, color='r', marker='o', linewidth=0)

pyplot.streamplot(X, Y, u_pair, v_pair, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B');
```
