##Source distribution on an Airfoil

In flow around a symmetric airfoil at $0^{\circ}$ angle of attack:


* *Where is the point of maximum pressure?*

In the middle of the nose of the airfoil. In that point the velocity is zero then the pressure is maximum. 

* *What do we call that point?*

We call that point stagnation point

* *Will the airfoil generate any lift?*

Due to the symmetry, it won't generate any lift.

### Problem setup

We will read data files containing information about the location and the strength of a set of sources located on the surface of a NACA0012 airfoil.

Using 51 mesh points in each direction, and a domain $[-1, 2]\times[-0.5, 0.5]$, compute the velocity due to the set of sources plus a free stream in the $x$-direction with $U_{\infty}=1$. Also compute the coefficient of pressure on your grid points.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
x_s = numpy.loadtxt('resources/NACA0012_x.txt')
y_s = numpy.loadtxt('resources/NACA0012_y.txt')
sigma = numpy.loadtxt('resources/NACA0012_sigma.txt')
```

```python
N = 51                                    # Number of point in each direction

x_start, x_end = -1.0 , 2.0                # x-boundaries (length 4)
y_start, y_end = -0.5 , 0.5                # y-boundaries (length 2)
 
x = numpy.linspace(x_start, x_end, N)      # 1D array in x-direction equally spaced and with N points
y = numpy.linspace(y_start, y_end, N)      # 1D array in y-direction equally spaced and with N points

#Let's generate the mesh grid using x and y

X, Y = numpy.meshgrid(x, y)  
```

```python
#scatter for the source and contour for psi_stagnation

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#source and stagnation point
pyplot.scatter(x_s, y_s, s=30, color='b', marker='o', linewidth=0);
```

```python
#Function to get the velocity components of a sink or source

def get_velocity(strength, x_s, y_s,num_sources, X, Y):
    
    """Returns the velocity field (u, v) generated by a set of sources
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    u=numpy.zeros((N, N))
    v=numpy.zeros((N, N))
    
    for i in range(num_sources):
        
        u += strength[i]/(2*numpy.pi)*(X-x_s[i])/((X-x_s[i])**2+(Y-y_s[i])**2)
        v += strength[i]/(2*numpy.pi)*(Y-y_s[i])/((X-x_s[i])**2+(Y-y_s[i])**2)
    
    return u, v
```

```python
#Function to get the stream function of a source or sink

def get_stream_function(strength, x_s, y_s,num_sources, X, Y):
    
    """Returns the stream-function generated by a set of sources.
    
    Arguments
    ---------
    strength: array with the strength of the sources .
    x_s, y_s: arrays with the coordinates of the source.
    X, Y: mesh grid.
    """
    psi=numpy.zeros((N, N))
    
    for i in range(num_sources):
        
        psi += strength[i]/(2*numpy.pi)*numpy.arctan2((Y-y_s[i]), (X-x_s[i]))
    
    return psi
```

```python
num_sources = len(sigma)
```

```python
# computes the velocity field
u_sources, v_sources = get_velocity(sigma, x_s, y_s,num_sources, X, Y)

# computes the stream-function
psi_sources = get_stream_function(sigma, x_s, y_s, num_sources,X, Y)
```

```python
#Uniform flow velocities

U_inf = 1.0       #freestream speed

#Velocity field

u_freestream = U_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

#Stream-function

psi_freestream = U_inf * Y
```

```python
#Let's do superposition and plot the result

# superposition of the source on the freestream
u = u_freestream + u_sources
v = v_freestream + v_sources

psi = psi_freestream + psi_sources
```

```python
#Now plotting the streamlines using streamplot(), 
#scatter for the source and contour for psi_stagnation

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#sources
pyplot.scatter(x_s, y_s, s=30, color='r', marker='o', linewidth=0)

#streamlines
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1.2, arrowstyle='-|>',color='#483D8B');
```

###Bernoulli's equation and the pressure coefficient

From the previous notebooks, we know that through the Bernoulli's equation we can define the pressure coefficient as the ratio between the pressure difference with the free stream, and the dynamic pressure:

$$C_p = \frac{p-p_\infty}{\frac{1}{2}\rho U_\infty^2}$$

i.e.,

$$C_p = 1 - \left(\frac{U}{U_\infty}\right)^2$$

In an incompressible flow, $C_p=1$ at a stagnation point. Let's plot the pressure coefficient in the whole domain.

```python
# computes the pressure coefficient field
C_p = 1.0 - (u**2+v**2)/U_inf**2
```

```python
#Plotting C_p

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label


levels = numpy.linspace(numpy.min(C_p), numpy.max(C_p), 100)

#C_p field
pyplot.contourf(X, Y, C_p, levels)
cb = pyplot.colorbar()
cb.set_label('$C_p$', fontsize=16)
cb.set_ticks(numpy.linspace(numpy.min(C_p), numpy.max(C_p), 10))

#Airfoil
pyplot.scatter(x_s, y_s, color='k', s=20, marker='o');


```

1. What is the value of maximum pressure coefficient, $C_p$?
2. What are the array indices for the maximum value of $C_p$?

```python
#Maximum value of pressure coefficient
Cp_max=numpy.amax(C_p)

print ('Cp_max= %.6f' %Cp_max) 
```

```python
dim = C_p.shape
```

```python
#Get indices of whe C_p is maximum
loc_Cp_max = numpy.argmax(C_p)

Cp_max_indices = numpy.unravel_index(loc_Cp_max, dim)

print Cp_max_indices
```

* Distribution of the pressure coefficient and a single marker on the location of the maximum pressure

```python
#Using flat, we can get where is the maximum in the mesh.

X_Cp_max = X.flat[loc_Cp_max]
Y_Cp_max = Y.flat[loc_Cp_max]

print ('X_Cp_max= %.2f , Y_Cp_max= %.2f' %(X_Cp_max,Y_Cp_max))
```

```python
#Let's plot again, but now adding the point of maximum pressure and also
#The streamlines.

#Plotting C_p

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label


levels = numpy.linspace(numpy.min(C_p), numpy.max(C_p), 100)

#C_p field
pyplot.contourf(X, Y, C_p, levels)
cb = pyplot.colorbar()
cb.set_label('$C_p$', fontsize=16)
cb.set_ticks(numpy.linspace(numpy.min(C_p), numpy.max(C_p), 10))

#Airfoil
pyplot.scatter(x_s, y_s, color='k', s=20, marker='o')

#streamlines
pyplot.streamplot(X, Y, u, v, density=1, linewidth=0.5, arrowsize=1.2, arrowstyle='-|>',color='k');

#Point of maximum pressure (white star *)
pyplot.scatter(X_Cp_max, Y_Cp_max, color='w', s=60, marker='*');
```
