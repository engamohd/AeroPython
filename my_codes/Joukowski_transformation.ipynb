###Joukowski transformation

Consider a point in the complex plane:

$$\zeta=\xi+i\eta$$

mapped to another plane at:

$$z=x+iy$$

by the transformation:

$$z=\zeta+\frac{1}{\zeta}$$

This is called the Joukowski transformation, and it can map a circle to an airfoil.

Let's explore this transformation. Do the following:

* Write and test a function that takes any point or array and maps it from the $\zeta$ plane to the z plane.

1. Place a cylinder of radius 1 at the origin, map it ... plot the results.

2. Repeat with a cylinder of radius larger than 1, placed at the origin.

3. Now a cylinder of r=1.5, centered at (-0.5,0).

4. Get the value of streamfunction on a grid of points for flow around the cylinder of (3); now map those points to the $z$ plane ... plot the locations of the mapped points.

4. What happens if you try to plot the streamlines on the mapped plane?

```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Creating theta array
N_theta=300
theta = numpy.linspace(0, 2*numpy.pi, N_theta)
```

```python
def Joukowski_transformation(R, theta, x_c, y_c):
    
    """Returns the original coordinates (x,y) and
      their Joukowski transformation (x_jt, y_jt).
    
    Arguments
    ---------
    R: radius of the circle (float).
    theta: array of angles.
    x_c, y_c: x and y coordinates of the center of the circle .
    """
    zeta=R*numpy.exp(1j*theta) + (x_c + 1j*y_c)
    
    x=zeta.real
    y=zeta.imag
    
    z = zeta + 1./zeta
    
    x_jt=z.real
    y_jt=z.imag
  
    return x, y , x_jt, y_jt
```

```python
# cylinder of radius 1 at the origin

x_1, y_1 , x_jt_1, y_jt_1 = Joukowski_transformation(1., theta, 0., 0.)

# cylinder of radius > 1 at the origin

x_2, y_2 , x_jt_2, y_jt_2 = Joukowski_transformation(2.5, theta, 0., 0.)

# cylinder of radius 1.5 at (-0.5,0)

x_3, y_3 , x_jt_3, y_jt_3 = Joukowski_transformation(1.5, theta, -0.5, 0.)
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label
pyplot.title('JT of cylinder of radius 1 at the origin', fontsize=17)

#Plotting the circle
pyplot.scatter(x_1, y_1, color='g', s=20, alpha=0.1, marker='o')
pyplot.scatter(x_jt_1, y_jt_1, color='r', s=20, marker='o')

pyplot.legend(['cylinder', 'Joukowski transf.'], loc='best', prop={'size':15});
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label
pyplot.title('JT of cylinder of radius 2.5 at the origin', fontsize=17)

#Plotting the circle
pyplot.scatter(x_2, y_2, color='g', s=20, alpha=0.1, marker='o')
pyplot.scatter(x_jt_2, y_jt_2, color='r', s=20, marker='o')

pyplot.legend(['cylinder', 'Joukowski transf.'], loc='best', prop={'size':15});
```

Let's calculate the streamlines around the cylinder in case 3 and then mapped them in the $z$ plane using Joukowski transformation.

Remember that in cartesian coordinates the velocity components for the combined freestream + cylinder (doublet) are:

$$u\left(x,y\right) = \frac{\partial\psi}{\partial y} = U_\infty -\frac{\kappa}{2\pi}\frac{x^2-y^2}{\left(x^2+y^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\partial\psi}{\partial x} = -\frac{\kappa}{2\pi}\frac{2xy}{\left(x^2+y^2\right)^2}$$

But this components are valid when the cylinder is centered in $(0,0)$, if we have a shift cylinder then the velocity components are:

$$u\left(x,y\right) = U_\infty -\frac{\kappa}{2\pi}\frac{(x-x_c)^2-(y-y_c)^2}{\left((x-x_c)^2+(y-y_c)^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{2(x-x_c)(y-y_c)}{\left((x-x_c)^2+(y-y_c)^2\right)^2}$$

where $x_c$ and $y_c$ are the coordinates of the center of the cylinder.

Written in terms of the radius of the cylinder $R = \sqrt{\frac{\kappa}{2\pi U_\infty}}$ we have: 

$$u\left(x,y\right) = 1 -R^2\frac{(x-x_c)^2-(y-y_c)^2}{\left((x-x_c)^2+(y-y_c)^2\right)^2}$$

$$v\left(x,y\right) = -R^2\frac{2(x-x_c)(y-y_c)}{\left((x-x_c)^2+(y-y_c)^2\right)^2}$$


```python
#Let's create a mesh grid

N = 100                                   # Number of points in each direction
x_start, x_end = -5.0, 5.0               # x-direction boundaries
y_start, y_end = -3.0, 3.0                 # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
R_c = 1.5                           # radius of the cylinder
x_c, y_c = -0.5, 0.0                # location of the cylinder

U_inf = 1.0                         # freestream speed
```

```python
#Functions to get the velocity field

def get_velocity_cylinder_fs(R, x_cy, y_cy, X, Y):
    
    """Returns the velocity field generated by a
       cylinder in a freestream.
    
    Arguments
    ---------
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    u = 1-R**2*((X-x_cy)**2-(Y-y_cy)**2)/((X-x_cy)**2+(Y-y_cy)**2)**2
    v = -R**2*2*(X-x_cy)*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)**2
    
    return u, v
    
```

```python
def get_stream_function_cyl_fs(U_inf, R, x_cy, y_cy, X, Y):
    
    """Returns the stream-function generated by a cylinder
    in a freestream.
    
     Arguments
    ---------
    U_inf: freestream speed
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    psi = U_inf*(Y-y_c)- R**2*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)
    
    return psi
```

```python
# computes the velocity field of the cylinder on the mesh
u, v = get_velocity_cylinder_fs(R_c, x_c, y_c, X, Y)

# computes the stream-function on the mesh
psi = get_stream_function_cyl_fs(U_inf, R_c, x_c, y_c, X, Y)
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#Args: location (x,y), radius, color, transparency 
circle=pyplot.Circle((-0.5, 0), radius=R_c, color='#CD2305', alpha=0.5)
#gca() return the current axis, and add_patch(circle) says add the patch described in circle
pyplot.gca().add_patch(circle)

#streamlines
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1.2,\
                  arrowstyle='-|>',color='#483D8B');
```

```python
#Now plotting the potential lines using contour() 
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

#Args: location (x,y), radius, color, transparency 
circle=pyplot.Circle((-0.5, 0), radius=R_c, color='#CD2305', alpha=0.1)
#gca() return the current axis, and add_patch(circle) says add the patch described in circle
pyplot.gca().add_patch(circle)

levels = numpy.linspace(numpy.min(psi), numpy.max(psi), 300)

pyplot.contour(X, Y, psi, levels)
cb=pyplot.colorbar(shrink=0.8)
cb.set_ticks(numpy.linspace(numpy.min(psi), numpy.max(psi), 15))
```

Now we have the vector field, so let's map them into the $z$ plane using the Joukowski transformation.

```python
#Mapping the grid to the z plane

ZETA= X+1j*Y

Z = ZETA + 1./ZETA

X_z = Z.real
Y_z = Z.imag
```

```python
# And plotting the points

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x_z', fontsize=15)    #set x-label
pyplot.ylabel('y_z', fontsize=15)    #set y-label

pyplot.scatter(X_z, Y_z, s=10, color='g', marker='o', linewidth=0);
```

```python
#Now plotting the potential lines using contour() 
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_jt_3, y_jt_3, color='r', s=20, marker='o', alpha=0.2)

levels = numpy.linspace(numpy.min(psi), numpy.max(psi), 300)

pyplot.contour(X_z, Y_z, psi, levels)
cb=pyplot.colorbar(shrink=0.8)
cb.set_ticks(numpy.linspace(numpy.min(psi), numpy.max(psi), 15))
```
