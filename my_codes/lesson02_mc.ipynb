### Source-sink in a freestream

In this notebook we will solve the potential flow resulting from add a source, a sink, and a uniform free stream. We can do this due to the superposition principle (consequence of linearity of potential flow). 

```python
#Importing libraries and modules we need

import numpy
from matplotlib import pyplot
%matplotlib inline 
```

Let's creat the grid where we will compute the velocity. We did this in the previous notebook, so it shouldn't be difficult.

```python
N = 200                                    # Number of point in each direction

x_start, x_end = -4.0 , 4.0                # x-boundaries (length 4)
y_start, y_end = -2.0 , 2.0                # y-boundaries (length 2)
 
x = numpy.linspace(x_start, x_end, N)      # 1D array in x-direction equally spaced and with N points
y = numpy.linspace(y_start, y_end, N)      # 1D array in y-direction equally spaced and with N points

#Let's generate the mesh grid using x and y

X, Y = numpy.meshgrid(x, y)  
```

#### Source in a uniform flow

We'll first see what happend with a source in a uniform flow. We already known the streamfunction of a source but we need to know the one corresponding to a uniform flow.

The streamlines of a freestream with speed $U_\infty$ and angle of attack $\alpha$ are given by:

$$\psi_\text{freestream}(x,y) = U_\infty (y\cos\alpha - x\sin\alpha)$$

If the flow is completely horizontal ($\alpha=0 $i.e.parallel to x-direction), $\psi= U_\infty y$. Integrating, we get that $u=U_\infty$ and $v=0$.

Let's first write some code for the case of horizontal flow $(U_\infty,\alpha=0)$ and then we can try to try with a different $\alpha$.


```python
#Uniform flow velocities

U_inf = 1.0       #freestream speed

#Velocity field

u_freestream = U_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

#Stream-function

psi_freestream = U_inf * Y
```

For the source we can get all what we need from the previous lesson. However, it would be useful to write a function that give us the vector field. Morover, we can write a function that only changing the strength we can get the vector field of a source or a sink. (Remember both were the same except of the sign of the strength). 

```python
#Function to get the velocity components of a sink or source

def get_velocity(strength, x_s, y_s, X, Y):
    
    """Returns the velocity field (u, v) generated by a source/sink.
    
    Arguments
    ---------
    strength: strength of the source/sink.
    x_s, y_s: coordinates of the source/sink.
    X, Y: mesh grid.
    """
    u = strength/(2*numpy.pi)*(X-x_s)/((X-x_s)**2+(Y-y_s)**2)
    v = strength/(2*numpy.pi)*(Y-y_s)/((X-x_s)**2+(Y-y_s)**2)
    
    return u, v
```

We can also define a function to calculate the stream function of the source or sink on the mesh grid. First, remember that we can write the stream function of any of this singularities (source or sink) in cartesian coordinates as:

$$\psi_\text{source}(x,y) = \frac{\sigma}{2\pi} \arctan \left(\frac{y-y_\text{s}}{x-x_\text{s}}\right)$$

where $(x_s,y_s)$ is the location of the source (sink). 


```python
#Function to get the stream function of a source or sink

def get_stream_function(strength, x_s, y_s, X, Y):
    
    """Returns the stream-function generated by a source/sink.
    
    Arguments
    ---------
    strength: strength of the source/sink.
    x_s, y_s: coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    psi = strength/(2*numpy.pi)*numpy.arctan2((Y-y_s), (X-x_s))
    
    return psi
```

We use [`numpy.arctan2()`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.arctan2.html), check the documentation to be sure how it works.

```python
#Let's use these functions

strength_source = 5.0            # strength of the source
x_source, y_source = -1.0, 0.0   # location of the source

# computes the velocity field
u_source, v_source = get_velocity(strength_source, x_source, y_source, X, Y)

# computes the stream-function
psi_source = get_stream_function(strength_source, x_source, y_source, X, Y)
```

Now using  superposition we'll calculate the streamlines of the combination of a freestream and a source flow:

$$\psi = \psi_{freestream}+\psi_{source} = U_\infty y + \frac{\sigma}{2\pi} \arctan \left(\frac{y-y_\text{source}}{x-x_\text{source}}\right)$$

Due to differentiation is linear, the velocity field of the new flow pattern is:

$$u = u_\text{freestream} + u_\text{source}$$
$$v = v_\text{freestream} + v_\text{source}$$

The stagnation points in the flow are points where the velocity is zero. To find their location, we solve the following equations:

$$u = 0 , \quad v = 0$$

which leads in this case to:

$$x_\text{stagnation} = x_\text{source} - \frac{\sigma}{2\pi U_\infty}$$

$$y_\text{stagnation} = y_\text{source}$$
