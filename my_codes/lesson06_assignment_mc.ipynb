```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Parameters
c=1.
R_o=1.15
x_c, y_c = c-R_o, 0.

#Creating theta and r array
N_r = 100
r = numpy.linspace(R_o, 5., N_r)
N_theta=145
theta = numpy.linspace(0., 2*numpy.pi, N_theta)
```

```python
R, THETA = numpy.meshgrid(r, theta) 

Z = R*numpy.exp(1j*THETA) + (x_c + 1j*y_c)

X = Z.real
Y = Z.imag 

print Z.shape
print Z[62,0]
```

```python
#just the circle and the airfoil (to plot them after)
def JT_circle(R_c, theta, x_c, y_c):
    
    """Returns the original coordinates (x,y) and
      their Joukowski transformation (x_jt, y_jt).
    
    Arguments
    ---------
    c: constant parameter (usually 1.).
    theta: array of angles.
    x_c, y_c: x and y coordinates of the center of the circle .
    """
    z=R_c*numpy.exp(1j*theta) + (x_c + 1j*y_c)
    
    x_circle=z.real
    y_circle=z.imag
    
    zeta = z + 1./z
    
    x_airfoil=zeta.real
    y_airfoil=zeta.imag
  
    return x_circle, y_circle , x_airfoil, y_airfoil
```

```python
x_circle, y_circle , x_airfoil, y_airfoil = JT_circle(R_o, theta, x_c, y_c)
```

```python
def Joukowski_transformation_cyl(X,Y):
    
    """Returns the Joukowski transformation (X_z, Y_z).
    
    Arguments
    ---------
    X, Y : cartesian mesh points
    """        
    
    Z=X+1j*Y
    
    ZETA = Z + 1./Z
    
    X_zeta=ZETA.real
    Y_zeta=ZETA.imag
    
    return X_zeta, Y_zeta 
```

```python
X_zeta, Y_zeta = Joukowski_transformation_cyl(X,Y)
```

##Streamfunctions

```python
def get_stream_function_cyl_fs(strength, x_cy, y_cy, X, Y):
    
    """Returns the stream-function generated by a cylinder
    in a freestream.
    
     Arguments
    ---------
    U_inf: freestream speed
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    psi = U_inf*Y - strength/(2*numpy.pi)*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)
    
    return psi
```

```python
# computes the stream-function
U_inf = 1.0
kappa = R_o**2*2*numpy.pi*U_inf     # strength of the doublet
psi = get_stream_function_cyl_fs(kappa, x_c, y_c, X, Y)
```

```python
def get_velocity_doublet(strength, x_d, y_d, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    
    u = - strength/(2*numpy.pi)*((X-x_d)**2-(Y-y_d)**2)/((X-x_d)**2+(Y-y_d)**2)**2
    v = - strength/(2*numpy.pi)*2*(X-x_d)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)**2
    
    return u, v
```

```python
u_freestream = U_inf * numpy.ones_like(X)
v_freestream = numpy.zeros_like(Y)
```

```python
# Gamma = -U_inf*R_o*4.*numpy.pi*numpy.sin(-AoA_rad) , which is the same than:

Gamma = U_inf*R_o*4.*numpy.pi*numpy.sin(AoA_rad)
x_vortex , y_vortex = 0., 0. #In the z_prime 

print Gamma
```

```python
#Lif according Kutta-Joukowski Thm: y component is the component in y' cos(AoA)
Rho = 1.0
L = Rho * U_inf * Gamma * numpy.cos(AoA_rad)
print 'Lift according K-J Theorem =%.6f' %L
```

```python
def get_stream_function_vortex(strength, x_v, y_v, X, Y):
    
    """Returns the stream-function generated by a vortex.
    
    Arguments
    ---------
    strength: strength of the vortex.
    x_v, y_v: coordinates of the vortex.
    X, Y: mesh grid.
    """
    psi = strength/(4*numpy.pi)*numpy.log((X-x_v)**2+(Y-y_v)**2)
    
    return psi
```

```python
#Functions to get the velocity field and the stream function

def get_velocity_vortex(strength, x_v, y_v, X, Y):
    
    """Returns the velocity field generated by a vortex.
    
    Arguments
    ---------
    strength: strength of the vortex.
    x_v, y_v: coordinates of the vortex.
    X, Y: mesh grid.
    """
    u = + strength/(2*numpy.pi)*(Y-y_v)/((X-x_v)**2+(Y-y_v)**2)
    v = - strength/(2*numpy.pi)*(X-x_v)/((X-x_v)**2+(Y-y_v)**2)
    
    return u, v
```
