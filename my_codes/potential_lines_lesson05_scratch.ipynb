```python
#Function to get the potential lines
 
def get_potential_row_vortices(strength, x_v, y_v, X, Y):
    
    """Returns the stream-function generated by a vortex.
    
    Arguments
    ---------
    strength: strength of the vortex.
    x_v, y_v: coordinates of the vortex.
    X, Y: mesh grid.
    """
    phi=numpy.zeros((N, N))
    
    for i in range(len(x_v)):
        
        phi += -strength/(2*numpy.pi)*numpy.arctan2((Y-y_v[i]), (X-x_v[i]))
    
    return phi
```
