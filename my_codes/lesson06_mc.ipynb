#Lift on a cylinder

Is it possible for a perfectly circular cylinder to experience lift?
What if the cylinder is rotating?
All we need to do is add a vortex in the center of the cylinder. Let's see how that looks.


First, remember the equations for the flow of a doublet. In Cartesian coordinates, a doublet located at the origin has:

$$\psi\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}$$


$$u\left(x,y\right) = \frac{\partial\psi}{\partial y} = -\frac{\kappa}{2\pi}\frac{x^2-y^2}{\left(x^2+y^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\partial\psi}{\partial x} = -\frac{\kappa}{2\pi}\frac{2xy}{\left(x^2+y^2\right)^2}$$


##Let's compute

We'll place a doublet of strength $\kappa=1$ at the origin, and add a free stream $U_\infty=1$

```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Mesh-grid
N = 50                                   # Number of points in each direction
x_start, x_end = -2.0, 2.0               # x-direction boundaries
y_start, y_end = -1., 1.                 # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
kappa = 1.0                           # strength of the doublet
x_doublet, y_doublet = 0.0, 0.0       # location of the doublet

U_inf = 1.0                           # freestream speed
```

```python
#Functions for the doublet

def get_velocity_doublet(strength, x_d, y_d, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength: strength of the doublet.
    x_d, y_d: coordinates of the doublet.
    X, Y: mesh grid.
    """
    u = - strength/(2*numpy.pi)*((X-x_d)**2-(Y-y_d)**2)/((X-x_d)**2+(Y-y_d)**2)**2
    v = - strength/(2*numpy.pi)*2*(X-x_d)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)**2
    
    return u, v
```

```python
def get_stream_function_doublet(strength, x_d, y_d, X, Y):
    """Returns the stream-function generated by a doublet.
    
    Arguments
    ---------
    strength: strength of the doublet.
    x_d, y_d: coordinates of the doublet.
    X, Y: mesh grid.
    """
    psi = - strength/(2*numpy.pi)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)
    
    return psi
```

```python
#Doublet

# computes the velocity field on the mesh
u_doublet, v_doublet = get_velocity_doublet(kappa, x_doublet, y_doublet, X, Y)

# computes the stream-function on the mesh
psi_doublet = get_stream_function_doublet(kappa, x_doublet, y_doublet, X, Y)
```

```python
#Freestream

#Velocity field
u_freestream = U_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

# stream-function of the freestream flow
psi_freestream = U_inf * Y
```

```python
# superposition of the doublet on the freestream flow

u = u_freestream + u_doublet
v = v_freestream + v_doublet

psi = psi_freestream + psi_doublet
```

```python
#Plotting
pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

# calculates the cylinder radius and adds it to the figure
R = numpy.sqrt(kappa/(2*numpy.pi*U_inf)) #Notes in challange questions, lesson03
#Args: location (x,y), radius, color, transparency 
circle=pyplot.Circle((0, 0), radius=R, color='#CD2305', alpha=0.5)
#gca() return the current axis, and add_patch(circle) says add the patch described in circle
pyplot.gca().add_patch(circle)

#the stagnation points and add them to the figure
#The derivation is in my notebook
x_stagn1, y_stagn1 = +numpy.sqrt(kappa/(2*numpy.pi*U_inf)), 0
x_stagn2, y_stagn2 = -numpy.sqrt(kappa/(2*numpy.pi*U_inf)), 0

pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');

#Location of the doublet
pyplot.scatter(x_doublet, y_doublet, color='r', s=80, marker='o')

#streamlines
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1.2,\
                  arrowstyle='-|>',color='#483D8B');
```

Until the moment nothing new happens, so let's add a vortex located in the origin  $\Gamma$. In Cartesian coordinates,

$$\psi\left(x,y\right) = \frac{\Gamma}{4\pi}\ln\left(x^2+y^2\right)$$

$$u\left(x,y\right) = \frac{\Gamma}{2\pi}\frac{y}{x^2+y^2} \qquad v\left(x,y\right) = -\frac{\Gamma}{2\pi}\frac{x}{x^2+y^2}$$

We'll use the same functions than in lesson 04 and see what happens. 

```python
Gamma = 4.0                      # strength of the vortex
x_vortex, y_vortex = 0.0, 0.0    # location of the vortex
```

```python
#Functions to get the velocity field and the stream function

def get_velocity_vortex(strength, x_v, y_v, X, Y):
    
    """Returns the velocity field generated by a vortex.
    
    Arguments
    ---------
    strength: strength of the vortex.
    x_v, y_v: coordinates of the vortex.
    X, Y: mesh grid.
    """
    u = + strength/(2*numpy.pi)*(Y-y_v)/((X-x_v)**2+(Y-y_v)**2)
    v = - strength/(2*numpy.pi)*(X-x_v)/((X-x_v)**2+(Y-y_v)**2)
    
    return u, v
```

```python
def get_stream_function_vortex(strength, x_v, y_v, X, Y):
    
    """Returns the stream-function generated by a vortex.
    
    Arguments
    ---------
    strength: strength of the vortex.
    x_v, y_v: coordinates of the vortex.
    X, Y: mesh grid.
    """
    psi = strength/(4*numpy.pi)*numpy.log((X-x_v)**2+(Y-y_v)**2)
    
    return psi
```

```python
# computes the velocity field on the mesh grid
u_vortex, v_vortex = get_velocity_vortex(Gamma, x_vortex, y_vortex, X, Y)

# computes the stream-function on the mesh grid
psi_vortex = get_stream_function_vortex(Gamma, x_vortex, y_vortex, X, Y)
```

###Invoking superposition

Remember we previous call $u$ and $v$, the superposition of the doublet and free stream. So now, we should add the vortex velocities to $u$ and $v$.

```python
# superposition of the doublet and the vortex on the freestream flow

u_tot = u + u_vortex
v_tot = v + v_vortex
psi_tot = psi + psi_vortex
```

```python
#Plotting

pyplot.figure(figsize=(10,5))      #set plot size
pyplot.grid(True)                  #plot the grid

pyplot.xlim(x_start, x_end)        #set x-lim
pyplot.ylim(y_start, y_end)        #set y-lim
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

# calculates the cylinder radius and adds it to the figure
R = numpy.sqrt(kappa/(2*numpy.pi*U_inf)) #Notes in challange questions, lesson03
#Args: location (x,y), radius, color, transparency 
circle=pyplot.Circle((0, 0), radius=R, color='#CD2305', alpha=0.5)
#gca() return the current axis, and add_patch(circle) says add the patch described in circle
pyplot.gca().add_patch(circle)

# calculates the stagnation points
x_stagn1, y_stagn1 = +numpy.sqrt(R**2-(Gamma/(4*numpy.pi*U_inf))**2), -Gamma/(4*numpy.pi*U_inf)
x_stagn2, y_stagn2 = -numpy.sqrt(R**2-(Gamma/(4*numpy.pi*U_inf))**2), -Gamma/(4*numpy.pi*U_inf)

pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');

#Location of the vortex
pyplot.scatter(x_vortex, y_vortex, color='r', s=80, marker='o')

#streamlines
pyplot.streamplot(X, Y, u_tot, v_tot, density=2, linewidth=1, arrowsize=1.2,\
                  arrowstyle='-|>',color='#483D8B');
```
