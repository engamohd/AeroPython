```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Parameters
c=1.
x_c, y_c = -0.15, 0.
R=1.15

#Creating theta and r array
N_r = 100
r = numpy.linspace(R, 5., N_r)
N_theta=145
theta = numpy.linspace(0., 2*numpy.pi, N_theta)
```

```python
#just the circle and the airfoil (to plot them after)
def JT_circle(R_c, theta, x_c, y_c):
    
    """Returns the original coordinates (x,y) and
      their Joukowski transformation (x_jt, y_jt).
    
    Arguments
    ---------
    c: constant parameter (usually 1.).
    theta: array of angles.
    x_c, y_c: x and y coordinates of the center of the circle .
    """
    zeta=R_c*numpy.exp(1j*theta) + (x_c + 1j*y_c)
    
    x_circle=zeta.real
    y_circle=zeta.imag
    
    z = zeta + 1./zeta
    
    x_airfoil=z.real
    y_airfoil=z.imag
  
    return x_circle, y_circle , x_airfoil, y_airfoil
```

```python
x_circle, y_circle , x_airfoil, y_airfoil = JT_circle(R, theta, x_c, y_c)
```

```python
radius, angle = numpy.meshgrid(r,theta)

X = radius * numpy.cos(angle) + x_c
Y = radius * numpy.sin(angle) + y_c


#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X,Y,s=2, color='k', marker='o', linewidth=0);
```

```python
def Joukowski_transformation_cyl(X,Y):
    
    """Returns the Joukowski transformation (X_z, Y_z).
    
    Arguments
    ---------
    X, Y : cartesian mesh points
    """        
    
    ZETA=X+1j*Y
    
    Z = ZETA + 1./ZETA
    
    X_z=Z.real
    Y_z=Z.imag
    
    return X_z, Y_z 
```

```python
X_z, Y_z = Joukowski_transformation_cyl(X,Y)
```

```python
#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X_z,Y_z,s=2, color='k', marker='o', linewidth=0);
```

##Streamfunctions

```python
def get_stream_function_cyl_fs(R, x_cy, y_cy, X, Y):
    
    """Returns the stream-function generated by a cylinder
    in a freestream.
    
     Arguments
    ---------
    U_inf: freestream speed
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    psi = Y - R**2*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)
    
    return psi
```

```python
# computes the stream-function

psi = get_stream_function_cyl_fs(R, x_c, y_c, X, Y)
```

```python
#Now plotting the potential lines using contour() 
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlim(-6, 6)        #set x-lim
pyplot.ylim(-6, 6)        #set y-lim$

pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

levels = numpy.linspace(numpy.min(psi), numpy.max(psi), 50)

#pyplot.contour(X, Y, psi, levels)
pyplot.contour(X, Y, psi, levels, colors='k', linewidths=1, linestyles='solid')
pyplot.scatter(x_circle,y_circle,s=2, color='r', marker='o', linewidth=0)
#pyplot.contour(X, Y, psi, levels=[0.], colors='r', linewidths=2, linestyles='solid')

#cb=pyplot.colorbar(shrink=0.8)
#cb.set_ticks(numpy.linspace(numpy.min(psi), numpy.max(psi), 15));
```
