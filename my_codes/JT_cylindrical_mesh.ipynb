```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Parameters
c=1.
x_c, y_c = -0.15, 0
R=1.15

#Creating theta and r array
N_r = 100
r = numpy.linspace(R, 5., N_r)
N_theta=145
theta = numpy.linspace(0, 2*numpy.pi, N_theta)
```

```python
radius, angle = numpy.meshgrid(r,theta)

X = radius * numpy.cos(angle) + x_c
Y = radius * numpy.sin(angle) + y_c


#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X,Y,s=2, color='k', marker='o', linewidth=0);
```

```python
def Joukowski_transformation_cyl(X,Y):
    
    """Returns the Joukowski transformation (X_z, Y_z).
    
    Arguments
    ---------
    X, Y : cartesian mesh points
    """        
    
    ZETA=X+1j*Y
    
    Z = ZETA + 1./ZETA
    
    X_z=Z.real
    Y_z=Z.imag
    
    return X_z, Y_z 
```

```python
X_z, Y_z = Joukowski_transformation_cyl(X,Y)
```

```python
#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X_z,Y_z,s=2, color='k', marker='o', linewidth=0);
```

##Streamfunctions

```python
def get_stream_function_cyl_fs(U_inf, R, x_cy, y_cy, X, Y):
    
    """Returns the stream-function generated by a cylinder
    in a freestream.
    
     Arguments
    ---------
    U_inf: freestream speed
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    psi = U_inf*(Y-y_cy)- R**2*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)
    
    return psi
```

```python
U_inf = 1.

# computes the stream-function on both meshes

psi = get_stream_function_cyl_fs(U_inf, R, x_c, y_c, X, Y)

psi_jt = get_stream_function_cyl_fs(U_inf, R, x_c, y_c, X_z, Y_z)
```
