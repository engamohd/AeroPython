```python
#Importing the libraries and modules we need.
import numpy
from matplotlib import pyplot
#The next line allows as to plot directly in the notebook.
%matplotlib inline
```

```python
#Parameters
c=1.
R_o=1.15
x_c, y_c = c-R_o, 0.

#Creating theta and r array
N_r = 100
r = numpy.linspace(R_o, 5., N_r)
N_theta=145
theta = numpy.linspace(0., 2*numpy.pi, N_theta)
```

```python
R, THETA = numpy.meshgrid(r, theta) 

Z = R*numpy.exp(1j*THETA) + (x_c + 1j*y_c)

X = Z.real
Y = Z.imag 

print Z.shape
print Z[62,0]
```

```python
#just the circle and the airfoil (to plot them after)
def JT_circle(R_c, theta, x_c, y_c):
    
    """Returns the original coordinates (x,y) and
      their Joukowski transformation (x_jt, y_jt).
    
    Arguments
    ---------
    c: constant parameter (usually 1.).
    theta: array of angles.
    x_c, y_c: x and y coordinates of the center of the circle .
    """
    z=R_c*numpy.exp(1j*theta) + (x_c + 1j*y_c)
    
    x_circle=z.real
    y_circle=z.imag
    
    zeta = z + 1./z
    
    x_airfoil=zeta.real
    y_airfoil=zeta.imag
  
    return x_circle, y_circle , x_airfoil, y_airfoil
```

```python
x_circle, y_circle , x_airfoil, y_airfoil = JT_circle(R_o, theta, x_c, y_c)
```

```python
#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X,Y,s=2, color='k', marker='o', linewidth=0);
```

```python
def Joukowski_transformation_cyl(X,Y):
    
    """Returns the Joukowski transformation (X_z, Y_z).
    
    Arguments
    ---------
    X, Y : cartesian mesh points
    """        
    
    Z=X+1j*Y
    
    ZETA = Z + 1./Z
    
    X_zeta=ZETA.real
    Y_zeta=ZETA.imag
    
    return X_zeta, Y_zeta 
```

```python
X_zeta, Y_zeta = Joukowski_transformation_cyl(X,Y)
```

```python
#Plotting
pyplot.figure(figsize=(6,6))      #set plot size
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(X_zeta,Y_zeta,s=2, color='k', marker='o', linewidth=0);
```

##Streamfunctions

```python
def get_stream_function_cyl_fs(strength, x_cy, y_cy, X, Y):
    
    """Returns the stream-function generated by a cylinder
    in a freestream.
    
     Arguments
    ---------
    U_inf: freestream speed
    R: radius of the cylinder.
    x_cy, y_cy: coordinates of the cylinder.
    X, Y: mesh grid.
    """
    psi = U_inf*Y - strength/(2*numpy.pi)*(Y-y_cy)/((X-x_cy)**2+(Y-y_cy)**2)
    
    return psi
```

```python
# computes the stream-function
U_inf = 1.0
kappa = R_o**2*2*numpy.pi*U_inf     # strength of the doublet
psi = get_stream_function_cyl_fs(kappa, x_c, y_c, X, Y)
```

```python
#Now plotting the potential lines using contour() 
pyplot.figure(figsize=(14,6))      #set plot size

pyplot.subplot(1,2,1)
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlim(-6, 6)        #set x-lim
pyplot.ylim(-6, 6)        #set y-lim$
pyplot.title('$z$ plane', fontsize=16)
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

levels = numpy.linspace(numpy.min(psi), numpy.max(psi), 50)

#pyplot.contour(X, Y, psi, levels)
pyplot.contour(X, Y, psi, levels, colors='k', linewidths=1, linestyles='solid')
pyplot.scatter(x_circle,y_circle,s=5, color='r', marker='o', linewidth=0)
#pyplot.contour(X, Y, psi, levels=[0.], colors='r', linewidths=2, linestyles='solid')

#cb=pyplot.colorbar(shrink=0.8)
#cb.set_ticks(numpy.linspace(numpy.min(psi), numpy.max(psi), 15))

pyplot.subplot(1,2,2)
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlim(-6, 6)        #set x-lim
pyplot.ylim(-6, 6)        #set y-lim
pyplot.title('$\zeta$ plane', fontsize=16)
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

levels = numpy.linspace(numpy.min(psi), numpy.max(psi), 50)

#pyplot.contour(X_z, Y_z, psi, levels)
pyplot.contour(X_zeta, Y_zeta, psi, levels, colors='k', linewidths=1, linestyles='solid')
pyplot.scatter(x_airfoil,y_airfoil,s=5, color='r', marker='o', linewidth=0)
#pyplot.contour(X_z, Y_z, psi,levels=[0], colors='r', linewidths=2, linestyles='solid');

#cb=pyplot.colorbar(shrink=0.8)
#cb.set_ticks(numpy.linspace(numpy.min(psi), numpy.max(psi), 15));
```

Let's get the velocity field and then using some math we can obtain the velocity field in the transform plane.  

```python
def get_velocity_doublet(strength, x_d, y_d, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    
    u = - strength/(2*numpy.pi)*((X-x_d)**2-(Y-y_d)**2)/((X-x_d)**2+(Y-y_d)**2)**2
    v = - strength/(2*numpy.pi)*2*(X-x_d)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)**2
    
    return u, v
```

```python
u_freestream = U_inf * numpy.ones_like(X)
v_freestream = numpy.zeros_like(Y)
```

```python
x_doublet, y_doublet = x_c, y_c    # location of the doublet
u_d, v_d = get_velocity_doublet(kappa, x_doublet, y_doublet, X, Y)

print kappa
```

```python
U = u_d + u_freestream
V = v_d + v_freestream
```

```python
#Let's get the velocity in the zeta plane

W_z = U - 1j*V

dzeta_dz = 1. - (c/Z)**2 
W_zeta =   W_z / dzeta_dz

#The first value of U_zeta and V_zeta we get a nan, cause it's a 0/0 but
#this value should be 0, so let's impose it.

U_zeta = W_zeta.real
V_zeta = -W_zeta.imag

#and we force [0,0]=0

U_zeta[0,0]=0.
V_zeta[0,0]=0.

#print U_zeta
```

```python
step=3
pyplot.figure(figsize=(14,6))      #set plot size

pyplot.subplot(1,2,1)
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlim(-3.0, 3.0)       #set x-lim
pyplot.ylim(-3.0, 3.0)          #set y-lim$
pyplot.title('$z$ plane', fontsize=16)
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_circle,y_circle,s=10, color='r', marker='o', linewidth=0, alpha=0.4)
pyplot.quiver(X[::step,::step], Y[::step,::step], U[::step,::step], V[::step,::step], linewidth=0.1, color = 'k', scale = 30)

pyplot.subplot(1,2,2)
pyplot.axis('equal')
pyplot.grid(True)                  #plot the grid
pyplot.xlim(-3.0, 3.0)       #set x-lim
pyplot.ylim(-3.0, 3.0)          #set y-lim$
pyplot.title('$\zeta$ plane', fontsize=16)
pyplot.xlabel('x', fontsize=15)    #set x-label
pyplot.ylabel('y', fontsize=15)    #set y-label

pyplot.scatter(x_airfoil,y_airfoil,s=10, color='r', marker='o', linewidth=0, alpha=0.4)
pyplot.quiver(X_zeta[::step,::step], Y_zeta[::step,::step], U_zeta[::step,::step], V_zeta[::step,::step], linewidth=0.15, color = 'k', scale = 30);
```

```python
#Pressure coefficient

Cp_z = 1.0 - (U**2 + V**2)/U_inf**2
Cp_zeta = 1.0 - (U_zeta**2 + V_zeta**2)/U_inf**2

#print  Cp_z.min()
#print Cp_z.max()

#print Cp_zeta[:,0]
```

##Flow over a symmetric Joukowski airfoil with non-zero angle of attack, without circulation

```python
#In our case Z'=Z.exp(-iAoA) (cause our Z involves the x_c and y_c)
AoA = 20.
AoA_rad = AoA*numpy.pi/180.
Z_prime = (Z - (x_c+1j*y_c))*numpy.exp(-1j*AoA_rad)

X_prime = Z_prime.real
Y_prime = Z_prime.imag
```

```python
#To use the same stream function, we need to set the variables 
#x_c and y_c = 0 cause in this frame, the coordinates are centered in the
#cylinder.
psi_prime = get_stream_function_cyl_fs(kappa, 0, 0, X_prime, Y_prime)
```

```python
def get_velocity_doublet_fs(strength, x_d, y_d, X, Y):
    """Returns the velocity field generated by a doublet.
    
    Arguments
    ---------
    strength -- strength of the doublet.
    xd, yd -- coordinates of the doublet.
    X, Y -- mesh grid.
    """
    
    u = U_inf*numpy.ones_like(X) - strength/(2*numpy.pi)*((X-x_d)**2-(Y-y_d)**2)/((X-x_d)**2+(Y-y_d)**2)**2
    v = - strength/(2*numpy.pi)*2*(X-x_d)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)**2
    
    return u, v
```

```python
U_prime , V_prime = get_velocity_doublet_fs(kappa, 0., 0., X_prime, Y_prime)
```

```python
#Let's get the velocity in the z_prime plane

W_z_prime = U_prime - 1j*V_prime 
W_AoA_z = W_z_prime*numpy.exp(-1j*AoA_rad)

U_AoA = W_AoA_z.real
V_AoA = -W_AoA_z.imag

dzeta_dz = 1. - (c/Z)**2 
W_AoA_zeta =   W_AoA_z / dzeta_dz

U_AoA_zeta = W_AoA_zeta.real
V_AoA_zeta = -W_AoA_zeta.imag

#and we force [0,0]=0

U_AoA_zeta[0,0]=0.
V_AoA_zeta[0,0]=0.
```
